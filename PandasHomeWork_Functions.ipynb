```python
# Extracting information from DataFrame obout how many people in our DataFrame have normal BloodPressure. 
#Info - A normal blood pressure level is less than 120/80 mmHg

df['normal_BloodPressure'] = df['BloodPressure'].apply(lambda x: x > 80 and x < 120)
```
