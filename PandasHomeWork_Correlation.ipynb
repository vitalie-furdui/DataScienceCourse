```python
import numpy as np
import pandas as pd

# From library Kydavra importing the following FeatureSelection methods:
from kydavra import PearsonCorrelationSelector
from kydavra import SpearmanCorrelationSelector
from kydavra import KendallCorrelationSelector
from kydavra import PointBiserialCorrSelector
```

```python
#Upload DataFrame

df = pd.read_csv("data\diabetes.csv")
```

```python
#Printing DataFrame

df
```

```python
# Applying the Pearson correlation code

df.corr()
```

Coeficientul de corelaÈ›ie al lui Pearson este utilizat pentru a studia relaÈ›ia a douÄƒ variabile mÄƒsurate Ã®n scale metrice Ã®n setul de date. Ne permite sÄƒ determinam cÃ¢t de proporÈ›ionalÄƒ este variabilitatea Ã®ntre douÄƒ variabile.

```python
# Applying the Kendall correlation code

df.corr(method='kendall')
```

Coeficientul de corelaÈ›ie al lui Kendall calculeazÄƒ un scor normalizat pentru numÄƒrul de clasamente
coincidente sau consistente Ã®ntre douÄƒ eÈ™antioane.

```python
# Applying the Spearman correlation code

df.corr(method='spearman')
```

Coeficientul de corelaÈ›ie a rangului lui Spearman este utilizat pentru a identifica È™i a evalua strÃ¢nsitatea relaÈ›iei dintre douÄƒ serii de indicatori cantitativi comparaÈ›i.

```python
# Applying correlation code for columns from dataframe with column "Outcome" 

df.corrwith(df['Outcome'])
```

```python
# Applying correlation code with method by Kendall for columns from dataframe with column "Outcome"

df.corrwith(df['Outcome'], method='kendall')
```

```python
# Applying correlation code with method by Spearman for columns from dataframe with column "Outcome"

df.corrwith(df['Outcome'], method='spearman')
```

Prin aplicare metodelor de correlaÈ›ie, noi putem sÄƒ alegem caracteristici importante. Ca exemplu Ã®n fiÈ™ierul nostru nu toate coloanele sunt importante ca sÄƒ prezicem "Outcome". Noi putem din fiÈ™ierul nostru sÄƒ alegem cele mai Ã®mportante, prin aplicarea metodelor FeatureSelection din bibliotecÄƒ Kydavra.

```python
# Applying Pearson correlation code from library kydavra for dataframe with target column "Outcome" 

# Selectori bazaÈ›i pe correlaÈ›ie au 2 argumente max&min_corr. Un argument puternic pentru predicÈ›ie
# este atunci, cÃ®nd argumentul de correlaÈ›ie cu argumentul target este Ã®ntre 0.5 È™i 0.8. ÃŽn fiÈ™ierul
# nostru doar coloana "Glucose" este aproape de argumentul min_corr, si ca rezultat noi o sa primim
# valoare ce mai correlatÄƒ cu target "Glucose". È˜i Ã®n cazul dat eu am setat argumentul min_corr = 0.2.

pearson = PearsonCorrelationSelector(min_corr=0.2)
pearson.select(df, 'Outcome')
```

ÃŽn rezultatul dat noi vedem coloane care sunt cel mai puternic correlate cu coloana "Outcome" prin metoda Pearson.

```python
# Applying Spearman correlation code from library kydavra for dataframe with target column "Outcome"  

spearman = SpearmanCorrelationSelector(min_corr=0.2)
spearman.select(df, 'Outcome')
```

ÃŽn rezultatul dat noi vedem coloane care sunt cel mai puternic correlate cu coloana "Outcome" prin metoda Spearman.

```python
# Applying Kendall correlation code from library kydavra for dataframe with target column "Outcome" 

kendall = KendallCorrelationSelector(min_corr=0.2)
selected_cols = kendall.select(df, 'Outcome')
```

```python
selected_cols
```

ÃŽn rezultatul dat noi vedem coloane care sunt cel mai puternic correlate cu coloana "Outcome" 
prin metoda Kendall.

```python
# Printing dataframe of the strongest correlated argument to the target column "Outcome".

df[selected_cols + ['Outcome']]
```

```python
# Applying PointBiserialCorrSelector from library kydavra.

point_biserial = PointBiserialCorrSelector()
point_biserial.select(df, 'Outcome')
```
