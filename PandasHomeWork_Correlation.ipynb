```python
import numpy as np
import pandas as pd

# From library Kydavra importing the following FeatureSelection methods:
from kydavra import PearsonCorrelationSelector
from kydavra import SpearmanCorrelationSelector
from kydavra import KendallCorrelationSelector
from kydavra import PointBiserialCorrSelector
```

```python
#Upload DataFrame

df = pd.read_csv("data\diabetes.csv")
```

```python
#Printing DataFrame

df
```

```python
# Applying the Pearson correlation code

df.corr()
```

Coeficientul de corelație al lui Pearson este utilizat pentru a studia relația a două variabile măsurate în scale metrice în setul de date. Ne permite să determinam cât de proporțională este variabilitatea între două variabile.

```python
# Applying the Kendall correlation code

df.corr(method='kendall')
```

Coeficientul de corelație al lui Kendall calculează un scor normalizat pentru numărul de clasamente
coincidente sau consistente între două eșantioane.

```python
# Applying the Spearman correlation code

df.corr(method='spearman')
```

Coeficientul de corelație a rangului lui Spearman este utilizat pentru a identifica și a evalua strânsitatea relației dintre două serii de indicatori cantitativi comparați.

```python
# Applying correlation code for columns from dataframe with column "Outcome" 

df.corrwith(df['Outcome'])
```

```python
# Applying correlation code with method by Kendall for columns from dataframe with column "Outcome"

df.corrwith(df['Outcome'], method='kendall')
```

```python
# Applying correlation code with method by Spearman for columns from dataframe with column "Outcome"

df.corrwith(df['Outcome'], method='spearman')
```

Prin aplicare metodelor de correlație, noi putem să alegem caracteristici importante. Ca exemplu în fișierul nostru nu toate coloanele sunt importante ca să prezicem "Outcome". Noi putem din fișierul nostru să alegem cele mai împortante, prin aplicarea metodelor FeatureSelection din bibliotecă Kydavra.

```python
# Applying Pearson correlation code from library kydavra for dataframe with target column "Outcome" 

# Selectori bazați pe correlație au 2 argumente max&min_corr. Un argument puternic pentru predicție
# este atunci, cînd argumentul de correlație cu argumentul target este între 0.5 și 0.8. În fișierul
# nostru doar coloana "Glucose" este aproape de argumentul min_corr, si ca rezultat noi o sa primim
# valoare ce mai correlată cu target "Glucose". Și în cazul dat eu am setat argumentul min_corr = 0.2.

pearson = PearsonCorrelationSelector(min_corr=0.2)
pearson.select(df, 'Outcome')
```

În rezultatul dat noi vedem coloane care sunt cel mai puternic correlate cu coloana "Outcome" prin metoda Pearson.

```python
# Applying Spearman correlation code from library kydavra for dataframe with target column "Outcome"  

spearman = SpearmanCorrelationSelector(min_corr=0.2)
spearman.select(df, 'Outcome')
```

În rezultatul dat noi vedem coloane care sunt cel mai puternic correlate cu coloana "Outcome" prin metoda Spearman.

```python
# Applying Kendall correlation code from library kydavra for dataframe with target column "Outcome" 

kendall = KendallCorrelationSelector(min_corr=0.2)
selected_cols = kendall.select(df, 'Outcome')
```

```python
selected_cols
```

În rezultatul dat noi vedem coloane care sunt cel mai puternic correlate cu coloana "Outcome" 
prin metoda Kendall.

```python
# Printing dataframe of the strongest correlated argument to the target column "Outcome".

df[selected_cols + ['Outcome']]
```

```python
# Applying PointBiserialCorrSelector from library kydavra.

point_biserial = PointBiserialCorrSelector()
point_biserial.select(df, 'Outcome')
```
